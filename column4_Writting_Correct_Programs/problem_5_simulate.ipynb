```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (100,40)
```

```python
results = []
def Collatz_conjecture_simulation(number):
    round = 0
    while number != 1:
        if number%2==0:
            number/=2
        else:
            number = number*3 + 1
        round += 1
    results.append(round)
    
for n in range(1, 5000):
    Collatz_conjecture_simulation(n)
plt.plot(results)
```
